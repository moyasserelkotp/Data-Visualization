# Lesson: Multiple Plots

## Demonstration

Multiplots in Seaborn are powerful tools for visualizing relationships between multiple variables in a dataset. They help in identifying patterns, correlations, and distributions in a comprehensive manner. In this section, we will introduce two key multiplot functions: `pairplot` and `jointplot`.

The `pairplot` function creates a grid of plots, where each variable in the dataset is plotted against each other. It helps in visualizing the pairwise relationships and the distribution of each variable.

```python
# Load packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the tips dataset for this tutorial
tips = sns.load_dataset('tips')
```

```python
# Create a pairplot with hue based on the 'sex' attribute
sns.pairplot(tips, hue='sex')

# Show the plot
plt.show()
```

The `jointplot` function creates a bivariate plot with marginal histograms or density plots. It helps in visualizing the relationship between two variables, along with their individual distributions.

```python
# Create a jointplot for the relationship between 'total_bill' and 'tip'
sns.jointplot(data=tips , x='total_bill', y='tip', kind='scatter')

# Show the plot
plt.show()
```

```python
# Create a pairplot with regression lines and hue based on 'sex'
sns.pairplot(tips, hue='sex', kind='reg')

# Show the plot
plt.show()
```

```python
# Show the relationship between 'total_bill' and 'tip' with KDE
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')

# Show the plot
plt.show()
```
