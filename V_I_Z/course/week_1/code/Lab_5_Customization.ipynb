### Demonstration

Customizing plots involves modifying various elements like colors, line styles, markers, and more to improve clarity and aesthetics. 

**Syntax:**
- `plt.plot(x, y, color='color', linestyle='style', marker='marker', label='Label')`: Customizes the line plot.

Here's an overview of the customization options we'll explore:
- Color (`color`): Defines the color of the line. You can use color names (`'blue'`, `'green'`, `'red'`), hexadecimal codes (e.g., `'#FF5733'`), or single-letter abbreviations (e.g., `'b'` for blue).
- Line Style (`linestyle`): Specifies the style of the line. Common options include:
    - `'-'` for a solid line (default)
    - `'--'` for a dashed line
    - `':'` for a dotted line
    - `'-.'` for a dash-dot line
- Marker (`marker`): Adds symbols at data points to emphasize them. Examples include:
    - `'o'` for a circle
    - `'s'` for a square
    - `'x'` for a cross
    - `'^'` for an upward-pointing triangle
- Label (`label`): Specifies the label for a data series, displayed in the legend.


```python
import matplotlib.pyplot as plt 

# Sample data
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]

# Plotting the data with customizations
plt.plot(x, y1, color='g', linestyle='--', marker='o', label='Square')
plt.plot(x, y2, color='b', linestyle='-', marker='x', label='Linear')

# Adding title, labels, legend, and grid
plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Your Code Here
import matplotlib.pyplot as plt 
import numpy as np

dv1 = np.linspace(0,1,10)
dv2 = np.sqrt(dv1)
dv3 = dv1 * 2

# Plotting the data with customizations
plt.plot(dv1, dv2, color='r', linestyle=':', marker='d', label='Square')
plt.plot(dv1, dv3, color='b', linestyle='-', marker='x', label='Sqare Root')

# Adding title, labels, legend, and grid
plt.title('Square Vs SQRT')
plt.xlabel('X')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
