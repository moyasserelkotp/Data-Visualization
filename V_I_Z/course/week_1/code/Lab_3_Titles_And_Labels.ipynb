## Demonstration

Adding titles and axis labels helps interpret what the plot represents. For example, titles describe the plot as a whole, while axis labels clarify the data on each axis.


**Syntax:**
- `plt.title('Title')`: Adds a title to the plot.
- `plt.xlabel('Label')`: Adds a label to the x-axis.
- `plt.ylabel('Label')`: Adds a label to the y-axis.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Plotting the data
plt.plot(x, y)

# Adding title and labels
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```
