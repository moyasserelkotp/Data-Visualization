## Demonstration



Annotations allow you to highlight specific data points or add explanations directly to the plot.



**Syntax:**
- `plt.annotate('Text', xy=(x, y), xytext=(x_text, y_text), arrowprops=dict())`: Adds an annotation with an arrow pointing to the specified data point.
    - `'Text'`: The annotation text to display.
    - `xy=(x, y)`: The coordinates of the point to annotate.
    - `xytext=(x_text, y_text)`: The position where the annotation text will appear.
    - `arrowprops=dict()`: Properties for the arrow that connects the text to the point.
    
    
### Arrow Properties (`arrowprops=dict(`))
The arrowprops dictionary customizes the appearance of the arrow connecting the annotation text to the point:

- `facecolor`: Sets the arrow's color (e.g., `'black'`, `'red'`).
- `arrowstyle`: Specifies the arrow shape. Common styles include:
    - `'->'`: Standard arrow pointing to the data point.
    - `'-['`: Bracket-style arrow for emphasis.
    - `'<-'`: Arrow pointing away from the annotation text.
    - `'|-|'`: Bar at both ends of the line.
- `shrink`: Adjusts the size of the arrow. For example:
    - `shrink=0.05`: Shortens the arrow slightly.
    - `shrink=0.0`: Ensures the arrow reaches the exact point.

```python
import matplotlib.pyplot as plt 
# Sample data
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 16, 15, 12]

# Plotting the data
plt.plot(x, y)

# Adding title, labels, and annotation
plt.title('Annotated Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.annotate('Highest point', xy=(4, 16), xytext=(2, 15.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Display the plot
plt.show()
```
