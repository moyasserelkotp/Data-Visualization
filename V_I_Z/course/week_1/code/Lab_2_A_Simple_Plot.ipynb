## Demonstration

Creating a simple line plot is the first step to get familiar with Matplotlib. A line plot helps visualize the relationship between two variables, with one plotted along the x-axis and the other along the y-axis.

**Syntax:** `plt.plot(x, y)`
- `x`: The data for the x-axis.
- `y`: The data for the y-axis.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Plotting the data
plt.plot(x, y)

# Display the plot
plt.show()
```
