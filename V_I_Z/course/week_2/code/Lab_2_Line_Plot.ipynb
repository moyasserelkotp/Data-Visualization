# Lesson: Line Plot
Welcome to the lab on line plots! Line plots are one of the simplest and most commonly used visualizations for displaying trends over intervals. They help you understand the relationship between two variables, such as time and temperature, sales over months, or growth over years.

## Demonstration

A line plot is useful for displaying data trends over intervals. It shows information as a series of data points connected by straight line segments.

### Syntax
```python
plt.plot(x, y, options)
```
where:
- `x` represents the values along the x-axis.
- `y` represents the values along the y-axis.
- `options` allows customization such as marker style, line style, and color.

The following line plot shows the temperature variation over five days. Adding a title and labels helps understand what the axes represent.

If you would like to learn more about creating line plots with Matplotlib, you can visit the [Matplotlib Line Plot Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

````python
import matplotlib.pyplot as plt

# Dummy data
days = [1, 2, 3, 4, 5]
temperatures = [22, 23, 21, 25, 24]

# Creating the line plot
plt.plot(
    days,
    temperatures,
    color="b",
    linestyle="-",
    marker="o"
)

# Adding title and labels
plt.title('Temperature Over Days')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')

# Show plot
plt.show()
````
