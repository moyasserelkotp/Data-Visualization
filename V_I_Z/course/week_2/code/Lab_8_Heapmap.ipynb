Welcome to the lab on heatmaps! A heatmap visualizes data in a grid format where values are represented by colors. It is ideal for showing correlations or other matrix-based relationships. Let's get started!

## Demonstration

Heatmaps are useful for visualizing matrix-like data. They provide a clear visual representation of data density, patterns, and correlations.

### Syntax
The basic syntax for creating a heatmap in Matplotlib is:
```python
plt.imshow(data, options)
```

where: 
- `data` is the matrix to be visualized.
- `options` include colormap (cmap), interpolation, and other styling options.

The following heatmap represents random data values in a matrix format. Different colors represent different data ranges.

If you would like to learn more about creating heatmaps with Matplotlib, you can visit the [Matplotlib Heatmap Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html).

````python
import matplotlib.pyplot as plt
import numpy as np

# Dummy data
matrix = np.random.rand(10,10)

# Creating the heatmap
plt.imshow(matrix, cmap='viridis')

# Adding title
plt.title('Heatmap of Random Data')

# Show plot
plt.show()
````
