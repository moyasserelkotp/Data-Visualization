# Lesson: Pie Chart
Welcome to the lab on pie charts! Pie charts break data into proportional slices, making them perfect for representing parts of a whole.

## Demonstration

Pie charts are useful for showing the proportion of different categories. They provide a visual representation of how each category contributes to the whole.

### Syntax 
```python 
plt.pie(data, options)
```
where:
- `data` is the size of each slice.
- `options` include labels, colors, and percentage formatting.

The following pie chart shows the proportion of different fruits in a dataset. The `autopct` parameter displays the percentage of each category.

If you would like to learn more about creating pie charts with Matplotlib, you can visit the [Matplotlib Pie Chart Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html).

````python
import matplotlib.pyplot as plt

# Dummy data
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [15, 30, 45, 10]

# Creating the pie chart
plt.pie(sizes, labels=labels, colors=['r', 'y', 'c', 'm'])

# Adding title
plt.title('Fruit Proportions')

# Show plot
plt.show()
````
