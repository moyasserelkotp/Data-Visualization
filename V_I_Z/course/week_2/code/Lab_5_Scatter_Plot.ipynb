# Lesson: Scatter Plot
Welcome to the lab on scatter plots! Scatter plots help visualize relationships between two variables. Each data point represents one observation, making it easy to spot correlations or clusters. Let's get started!

## Demonstration

Scatter plots are useful for showing the relationship between two variables. They can help identify patterns, trends, and potential correlations between the variables.

### Syntax 
```python
plt.scatter(x, y, options)
``` 
where:
- `x` and `y` are the variables.
- `options` allows customization like color, size, and transparency.

The following scatter plot shows the relationship between the x and y variables. Each point represents a pair of values.

If you would like to learn more about creating scatter plots with Matplotlib, you can visit the [Matplotlib Scatter Plot Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).

````python
import matplotlib.pyplot as plt

# Dummy data
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Creating the scatter plot
plt.scatter(x, y, color='g')

# Adding title and labels
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')

# Show plot
plt.show()
````
