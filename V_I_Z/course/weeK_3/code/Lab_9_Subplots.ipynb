# Part 8: Subplots
This case study will walk you through how to visualize economic data using various Matplotlib plots. We will use the dataset provided, which contains the following columns:

- Year
- GDP
- GDP_PerCapita
- GDP_Growth
- Inflation
- Unemployment
- Inflation_Change

In this final part, we'll use Matplotlib to create subplots to visualize various economic indicators together. To begin, import the libraries and data in the cell below. Then, complete the following exercises.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('/content/EcoOfUs.csv')
df.info()
```

### Exercise 1
Create a figure with multiple subplots using `plt.subplots(nrows, ncols)` to visualize various economic indicators together.


```python
# Your Code Here
```

```python
# Create a 2x2 subplot
fig, axs = plt.subplots(2, 2)

# Subplot 1: GDP Over the Years
axs[0, 0].plot(df['Year'], df['GDP'], color='b')
axs[0, 0].set_title('GDP Over the Years')
axs[0, 0].set_xlabel('Year')
axs[0, 0].set_ylabel('GDP (in billions)')
axs[0, 0].grid(True)

# Subplot 2: GDP Growth Rate Over the Years
axs[0, 1].bar(df['Year'], df['GDP_Growth'], color='g')
axs[0, 1].set_title('GDP Growth Rate Over the Years')
axs[0, 1].set_xlabel('Year')
axs[0, 1].set_ylabel('GDP Growth Rate (%)')

# Subplot 3: Unemployment Rate Over the Years
axs[1, 0].plot(df['Year'], df['Unemployment'], color='r')
axs[1, 0].set_title('Unemployment Rate Over the Years')
axs[1, 0].set_xlabel('Year')
axs[1, 0].set_ylabel('Unemployment Rate (%)')
axs[1, 0].grid(True)

# Subplot 4: Inflation Over the Years
axs[1, 1].plot(df['Year'], df['Inflation'], color='m')
axs[1, 1].set_title('Inflation Over the Years')
axs[1, 1].set_xlabel('Year')
axs[1, 1].set_ylabel('Inflation (%)')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Subplot 1: GDP Over the Years
plt.subplot(2, 2, 1)
plt.plot(df['Year'], df['GDP'], color='b')
plt.title('GDP Over the Years')
plt.xlabel('Year')
plt.ylabel('GDP (in billions)')
plt.grid(True)

# Subplot 2: GDP Growth Rate Over the Years
plt.subplot(2, 2, 2)
plt.bar(df['Year'], df['GDP_Growth'], color='g')
plt.title('GDP Growth Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('GDP Growth Rate (%)')

# Subplot 3: Unemployment Rate Over the Years
plt.subplot(2, 2, 3)
plt.plot(df['Year'], df['Unemployment'], color='r')
plt.title('Unemployment Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)

# Subplot 4: Inflation Over the Years
plt.subplot(2, 2, 4)
plt.plot(df['Year'], df['Inflation'], color='m')
plt.title('Inflation Over the Years')
plt.xlabel('Year')
plt.ylabel('Inflation (%)')
plt.grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
