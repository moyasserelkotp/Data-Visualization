# Part 1: Line Plots

This case study will walk you through how to visualize economic data using various Matplotlib plots. We will use the dataset provided, which contains the following columns:
- Year
- GDP
- GDP_PerCapita
- GDP_Growth
- Inflation
- Unemployment
- Inflation_Change

In this first part, we'll use Matplotlib to create lineplots to visualize trends for GDP, GDP per Capita, and inflation. To begin, import the libraries and data in the cell below. Then, complete the following exercises.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('EcoOfUs.csv')
```

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.sample(5)
```

### Exercise 1
Create line plots to visualize trends over the years for GDP.

```python
# Your Code Here
plt.plot(df["Year"], df["GDP"])

plt.title("GDP vs Year")
plt.xlabel("Year")
plt.ylabel("GDP")

plt.show()
```

```python
# Line plot for GDP over the years
plt.plot(df['Year'], df['GDP'], marker='o')
plt.title('GDP Over the Years')
plt.xlabel('Year')
plt.ylabel('GDP (in billions)')
plt.grid()
plt.show()
```

### Exercise 2
Create line plots to visualize trends over the years for GDP Per Capita.

```python
# Your Code Here
plt.plot(df["Year"], df["GDP_PerCapita"])

plt.title("GDP Per Capita Over the Years")
plt.xlabel("Year")
plt.ylabel("GDP Per Capita")

plt.grid()
plt.show()
```

```python
# Line plot for GDP Per Capita over the years
plt.plot(df['Year'], df['GDP_PerCapita'], marker = 'o')
plt.title('GDP Per Capita Over the Years')
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')
plt.grid() 
plt.show()
```

### Exercise 3
Create line plots to visualize trends over the years for Inflation.

```python
# Your Code Here
plt.plot(df["Year"], df["Inflation"],marker = "o")
plt.title("GDP Per Capita Over the Years")
plt.xlabel("Year")
plt.ylabel("GDP Per Capita")
plt.grid()
plt.show()
```

```python
# Line plot for Inflation over the years
plt.plot(df['Year'], df['Inflation'], marker = 'o')
plt.title('Inflation Over the Years')
plt.xlabel('Year')
plt.ylabel('Inflation (%)')
plt.grid()
plt.show()
```
