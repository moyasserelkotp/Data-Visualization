# Part 2: Bar Plots

This case study will walk you through how to visualize economic data using various Matplotlib plots. We will use the dataset provided, which contains the following columns:

- Year
- GDP
- GDP_PerCapita
- GDP_Growth
- Inflation
- Unemployment
- Inflation_Change

In this second part, we'll use Matplotlib to create bar plots to compare GDP growth and unemployment over time. To begin, import the libraries and data in the cell below. Then, complete the following exercises.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("EcoOfUs.csv")
```

```python
df.info()
```

### Exercise 1
Create bar plots to compare GDP Growth across different years.


```python
# Your Code Here
plt.bar(df["Year"], df["GDP_Growth"], color="r")
plt.title("GDP Growth Rate Over the Years")
plt.xlabel("Year")
plt.ylabel("GDP Growth Rate (%)")
plt.show()
```

```python
# Bar plot for GDP Growth over the years
plt.bar(df['Year'], df['GDP_Growth'], color='g')
plt.title('GDP Growth Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('GDP Growth Rate (%)')
plt.show()
```

### Exercise 2 
Create bar plots to compare Unemployment across different years.


```python
# Your Code Here

plt.bar(df["Year"], df["Unemployment"], color="r")
plt.title("Unemployment Over the Years")
plt.xlabel("Year")
plt.ylabel("Unemployment (%)")
plt.show()
```

```python
# Bar plot for Unemployment over the years
plt.bar(df['Year'], df['Unemployment'], color='orange')
plt.title('Unemployment Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```
