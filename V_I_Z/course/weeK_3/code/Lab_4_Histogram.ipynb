# Part 3: Histogram
This case study will walk you through how to visualize economic data using various Matplotlib plots. We will use the dataset provided, which contains the following columns:

- Year
- GDP
- GDP_PerCapita
- GDP_Growth
- Inflation
- Unemployment
- Inflation_Change

In this third part, we'll use Matplotlib to create histograms to visualize the distribution of inflation rates and unemployement rates. To begin, import the libraries and data in the cell below. Then, complete the following exercises.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('EcoOfUs.csv')
```

```python
df.info()
```

### Exercise 1
Create a histogram to visualize the distribution of inflation rates.

```python
# Your Code Here

plt.hist(df["Inflation"],bins=10,edgecolor="black")

plt.title("Inflation Rate Over the Years")
plt.xlabel("Frequency")
plt.ylabel("Inflation Rate (%)")
plt.show()
```

```python
# Distribution of inflation rates
plt.hist(df['Inflation'], bins=10, edgecolor='black')
plt.title('Distribution of Inflation Rates')
plt.xlabel('Inflation Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

### Exercise 2
Create a histogram to visualize the distribution of unemployment rates.
