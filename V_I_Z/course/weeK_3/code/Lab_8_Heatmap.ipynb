```python
# Compute the correlation matrix
corr_matrix = df[['GDP', 'GDP_PerCapita', 'GDP_Growth', 'Inflation', 'Unemployment']].corr()

# Create a heatmap
plt.imshow(corr_matrix)
plt.title('Heatmap of Correlation Between Economic Indicators')
plt.show()
```
