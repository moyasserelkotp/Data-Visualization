# Part 4: Scatter Plots
This case study will walk you through how to visualize economic data using various Matplotlib plots. We will use the dataset provided, which contains the following columns:

- Year
- GDP
- GDP_PerCapita
- GDP_Growth
- Inflation
- Unemployment
- Inflation_Change

In this fourth part, we'll use Matplotlib to create scatter plots to visualize relationships between GDP and GDP per capita and between inflation and unemployment. To begin, import the libraries and data in the cell below. Then, complete the following exercises.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('EcoOfUs.csv')
```

```python
df.info()
```

### Exercise 1
Create scatter plots to visualize relationships between GDP and GDP Per Capita.

```python
# Your Code Here

plt.scatter(df["GDP"], df["GDP_PerCapita"])

plt.title("GDP vs. GDP Per Capita")
plt.xlabel("GDP (in billions)")
plt.ylabel("GDP Per Capita")
plt.show()
```

```python
# Scatter plot for GDP vs. GDP Per Capita
plt.scatter(df['GDP'], df['GDP_PerCapita'])

plt.title('GDP vs. GDP Per Capita')
plt.xlabel('GDP (in billions)')
plt.ylabel('GDP Per Capita')
plt.grid(True)
plt.show()
```

### Exercise 2
Create scatter plots to visualize relationships between Inflation and Unemployment.

```python
# Your Code Here
plt.scatter(df["Inflation"], df["Unemployment"])
plt.title("GDP vs. GDP Per Capita")
plt.xlabel("GDP (in billions)")
plt.ylabel("GDP Per Capita")
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Inflation vs. Unemployment
plt.scatter(df['Inflation'], df['Unemployment'])
plt.title('Inflation vs. Unemployment')
plt.xlabel('Inflation (%)')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```
